### Calculate how much is the final price that a customer will have to pay on a purchase, after the store offers the 15% discount

````python
price = float(input("Type the price $"))
discount = price * 0.15
FinalPurchasePrice = price - discount

print (f"The final price to pay is ${FinalPurchasePrice:.2f} " )
````
